# Setting out rules per column

```python
#Create a DataFrame object
#df_drugs = pd.DataFrame(data, columns=columns) 
r=60393
# Initiate df with True/False random values first
df_drugs = pd.DataFrame(np.random.choice([True, False],size=(r, len(columns))), columns=columns)

# Case ID: starting with 10,000 and plus number of cases available
n=10000
df_drugs.loc[:, 'Case ID'] = range(n,n+len(df_drugs))

# Alleged Crime: Set up the crime distribution, with Drug trafficking occuring 97% of the time and
# Drug possession occuring 3% of the time
df_drugs.loc[:, 'Alleged Crime'] = np.random.choice(
    alleged_crime, size=(len(df_drugs), 1), p = [0.97,0.03])

# Assign crime category:
df_drugs.loc[df_drugs['Alleged Crime'] == 'Drug trafficking', 'Alleged Crime Category'] = 'Felony'
df_drugs.loc[df_drugs['Alleged Crime'] == 'Drug possession', 'Alleged Crime Category'] = 'Misdemeanor' 

# Race: Number of different races are equally distributed across both alleged crimes
df_drugs.loc[:, 'Race'] = np.resize(random.choices(race, k=len(df_drugs)),len(df_drugs))

# Gender: 2% of cases are Non-U.S. Citizen
df_drugs.loc[:, 'Gender'] = np.random.choice(gender,size=(len(df_drugs), 1), p = [0.70,0.30])

# Citizenship: 2% of cases are Non-U.S. Citizen
df_drugs.loc[:, 'Citizenship'] = np.random.choice(citizenship,size=(len(df_drugs), 1), p = [0.98,0.02])

# Alleged Crime Reported
df_drugs.loc[:, 'Alleged Crime Reported'] = alleged_crime_reported

# Investigation:
# ['Unsolved or not arrested OUT', 'Arrest']
df_drugs.loc[:, 'Investigation'] = np.random.choice(investigation, size=(len(df_drugs), 1), p = [0.35,0.65])
# If after the investigation the case is OUT, make sure the rest of the values in the steps are set to NA
df_drugs.loc[df_drugs['Investigation'].str.contains('OUT'), 'Arrest':'Sentencing'] = 'NA'


### BIAS
def stage_bias(stage_name, stage_list, one_race, probability):
    stage_len = df_drugs.loc[(df_drugs[stage_name] != 'NA') & (df_drugs['Race'] == one_race), stage_name] 
    df_drugs.loc[(df_drugs[stage_name] != 'NA') & (
        df_drugs['Race'] == one_race), stage_name] = np.random.choice(stage_list, size=(len(stage_len)), p = probability)

# Arrest:
# ['Released OUT', 'Charges filed']
stage_bias("Arrest", arrest, "White", [0.40,0.60])
stage_bias("Arrest", arrest, "Black", [0.20,0.80])
stage_bias("Arrest", arrest, "Hispanic", [0.30,0.70])
stage_bias("Arrest", arrest, "Other", [0.30,0.70])
# If after the arrest the case is OUT, make sure the rest of the values in the steps are set to NA
df_drugs.loc[df_drugs['Arrest'].str.contains('OUT'), 'Charges filed':'Sentencing'] = 'NA'


# Charges filed: assign arrest outcomes to those that don't have NA
# ['Diversion', 'Court initial appearance']
### BIAS
stage_bias("Charges filed", charges_filed, "White", [0.20, 0.80])
stage_bias("Charges filed", charges_filed, "Black", [0.20, 0.80])
stage_bias("Charges filed", charges_filed, "Hispanic", [0.20, 0.80])
stage_bias("Charges filed", charges_filed, "Other", [0.20, 0.80])

# Diversion: assign diversion outcomes to those where Charges filed led to Diversion
# ['Released OUT', 'Court initial appearance'] 
s_diversion = df_drugs.loc[(df_drugs['Charges filed'] == 'Diversion') & (df_drugs['Diversion'] != 'NA'), 'Diversion'] 
df_drugs.loc[(df_drugs['Charges filed'] == 'Diversion') & (df_drugs['Diversion'] != 'NA'), 'Diversion'] = np.random.choice(diversion, size=(len(s_diversion), 1), p = [0.20,0.80])
# If Charges filed did not lead to a Diversion, fill out the column with NA
df_drugs.loc[df_drugs['Charges filed'] != 'Diversion', 'Diversion']  = 'NA'
# If after the diversion the case is OUT, make sure the rest of the values in the steps are set to NA
df_drugs.loc[df_drugs['Diversion'].str.contains('OUT'), 'Court initial appearance':'Sentencing'] = 'NA'

# Court initial appearance:
# ['Charges dropped or dismissed OUT', 'Move to preliminary hearing']
s_court_initial_appearance = df_drugs.loc[(df_drugs['Charges filed'] == 'Court initial appearance') | (df_drugs['Diversion'] == 'Court initial appearance') , 'Court initial appearance']
df_drugs.loc[(df_drugs['Charges filed'] == 'Court initial appearance') | (df_drugs['Diversion'] == 'Court initial appearance') , 'Court initial appearance'] = np.random.choice(court_initial_appearance, size=(len(s_court_initial_appearance), 1), p = [0.10,0.90])
# If after the court initial appearance the case is OUT, make sure the rest of the values in the steps are set to NA
df_drugs.loc[df_drugs['Court initial appearance'].str.contains('OUT'), 'Preliminary hearing':'Sentencing'] = 'NA'

# Preliminary hearing:
# ['Charges dropped or dismissed OUT', 'Bail or detention hearing']
s_preliminary_hearing = df_drugs.loc[df_drugs['Court initial appearance'] == 'Move to preliminary hearing', 'Preliminary hearing']
df_drugs.loc[df_drugs['Court initial appearance'] == 'Move to preliminary hearing', 'Preliminary hearing'] = np.random.choice(preliminary_hearing, size=(len(s_preliminary_hearing), 1), p = [0.10,0.90])
# If after the preliminary hearing the case is OUT, make sure the rest of the values in the steps are set to NA
df_drugs.loc[df_drugs['Preliminary hearing'].str.contains('OUT'), 'Bail or detention hearing':'Sentencing'] = 'NA'

# Bail or detention hearing:
# ['Move to grand jury', 'Move to collect information']
s_bail_detention_hearing = df_drugs.loc[df_drugs['Preliminary hearing'] == 'Bail or detention hearing', 'Bail or detention hearing']
df_drugs.loc[df_drugs['Preliminary hearing'] == 'Bail or detention hearing', 'Bail or detention hearing'] = np.random.choice(bail_detention_hearing, size=(len(s_bail_detention_hearing), 1), p = [0.002,0.998])

# Grand Jury:
# ['Refusal to indict OUT', 'Arraignment']
s_grand_jury = df_drugs.loc[df_drugs['Bail or detention hearing'] == 'Move to grand jury', 'Grand jury']
df_drugs.loc[df_drugs['Bail or detention hearing'] == 'Move to grand jury', 'Grand jury'] = np.random.choice(grand_jury, size=(len(s_grand_jury), 1), p = [0.005,0.995])
# If Bail or detention hearing did not lead to a Grand Jury:, fill out the column with NA
df_drugs.loc[df_drugs['Bail or detention hearing'] != 'Move to grand jury', 'Grand jury']  = 'NA'
# If after the grand jury the case is OUT, make sure the rest of the values in the steps are set to NA
df_drugs.loc[df_drugs['Grand jury'].str.contains('OUT'), 'Information stage':'Sentencing'] = 'NA'

# Information stage:
# ['Arraignment']
df_drugs.loc[df_drugs['Bail or detention hearing'] == 'Move to collect information', 'Information stage'] = information_stage
# If Bail or detention hearing did not lead to Information state:, fill out the column with NA
df_drugs.loc[df_drugs['Bail or detention hearing'] != 'Move to collect information', 'Information stage']  = 'NA'

# Arraignment
# ['Charges dismissed OUT','Plea bargain','Trial','Reduction of charge with trial']
s_arraignment = df_drugs.loc[(df_drugs['Grand jury'] == 'Arraignment') | (df_drugs['Information stage'] == 'Arraignment'), 'Arraignment'] 
df_drugs.loc[(df_drugs['Grand jury'] == 'Arraignment') | (df_drugs['Information stage'] == 'Arraignment'), 'Arraignment'] = np.random.choice(arraignment, size=(len(s_arraignment), 1), p = [0.05,0.70, 0.15, 0.10])
### BIAS
stage_bias("Arraignment", arraignment, "White", [0.04,0.88, 0.03, 0.05])
stage_bias("Arraignment", arraignment, "Black", [0.02,0.92, 0.04, 0.02])
stage_bias("Arraignment", arraignment, "Hispanic", [0.02,0.91, 0.04, 0.03])
stage_bias("Arraignment", arraignment, "Other", [0.02,0.91, 0.04, 0.03])

# If after the trial the case is OUT, make sure the rest of the values in the steps are set to NA
df_drugs.loc[df_drugs['Arraignment'].str.contains('OUT'), 'Plea bargain':'Sentencing'] = 'NA'

# Plea bargain
# ['Convicted']
df_drugs.loc[df_drugs['Arraignment'] == 'Plea bargain', 'Plea bargain'] = plea_bargain
# If Arraignment did not lead to a Plea Bargain:, fill out the column with NA
df_drugs.loc[df_drugs['Arraignment'] != 'Plea bargain', 'Plea bargain'] = 'NA'
df_drugs.loc[df_drugs['Arraignment'] == 'Plea bargain', 'Trial'] = 'NA'

# Trial
# ['Acquitted OUT', 'Convicted']
s_trial = df_drugs.loc[(df_drugs['Arraignment'] == 'Trial') | (df_drugs['Arraignment'] == 'Reduction of charge with trial'), 'Plea bargain']
df_drugs.loc[(df_drugs['Arraignment'] == 'Trial') | (df_drugs['Arraignment'] == 'Reduction of charge with trial'), 'Trial'] = np.random.choice(trial, size=(len(s_trial), 1), p = [0.05,0.95])
# If after the trial the case is OUT, make sure the rest of the values in the steps are set to NA
df_drugs.loc[df_drugs['Trial'].str.contains('OUT'), 'Convicted':'Sentencing'] = 'NA'

# Convicted
# ['Sentencing']
df_drugs.loc[(df_drugs['Plea bargain'] == 'Convicted') | (df_drugs['Trial'] == 'Convicted'), 'Convicted'] = convicted

# Sentencing
# ['Fine only', 'Prison only', 'Prison and alternatives', 'Probations only', 'Probation and alternatives']
stage_bias("Sentencing", sentencing, "White", [0.24,0.26, 0.20, 0.15, 0.15])
stage_bias("Sentencing", sentencing, "Black", [0.10,0.46, 0.34, 0.05, 0.05])
stage_bias("Sentencing", sentencing, "Hispanic", [0.10,0.40, 0.30, 0.10, 0.10])
stage_bias("Sentencing", sentencing, "Other", [0.10,0.40, 0.30, 0.10, 0.10])
```

### Additional sentencing data

```python
# Amount Range of Drug Possesed: [Currently proportionate by race]
# 1-17 *17 being the lowest range
# ['Investigation', 'Arrest', 'Charges filed', 'Diversion']
s_size = df_drugs.loc[df_drugs['NaN number'] >= 10, 'Amount Range of Drug Possesed']
df_drugs.loc[df_drugs['NaN number'] >= 10, 'Amount Range of Drug Possesed'] = np.random.choice(amount_in_possession[11:], size=len(s_size))

s_size = df_drugs.loc[(df_drugs['NaN number'] >= 6) & (df_drugs['NaN number'] <= 11), 'Amount Range of Drug Possesed']
df_drugs.loc[(df_drugs['NaN number'] >= 6) & (df_drugs['NaN number'] <= 11), 'Amount Range of Drug Possesed'] = np.random.choice(amount_in_possession[4:13], size=len(s_size))
    
s_size = df_drugs.loc[df_drugs['NaN number'] <= 6, 'Amount Range of Drug Possesed']
df_drugs.loc[df_drugs['NaN number'] <= 6, 'Amount Range of Drug Possesed'] = np.random.choice(amount_in_possession[:7], size=len(s_size)) 

# Severity of Crime:
# 1 least severe and 6 most severe
df_drugs.loc[(df_drugs['Amount Range of Drug Possesed'] == 1) | (df_drugs['Amount Range of Drug Possesed'] == 2), 'Severity of Crime'] = 6
df_drugs.loc[(df_drugs['Amount Range of Drug Possesed'] == 3) | (df_drugs['Amount Range of Drug Possesed'] == 4) | (df_drugs['Amount Range of Drug Possesed'] == 5), 'Severity of Crime'] = 5
df_drugs.loc[(df_drugs['Amount Range of Drug Possesed'] == 6) | (df_drugs['Amount Range of Drug Possesed'] == 7) | (df_drugs['Amount Range of Drug Possesed'] == 8), 'Severity of Crime'] = 4
df_drugs.loc[(df_drugs['Amount Range of Drug Possesed'] == 9) | (df_drugs['Amount Range of Drug Possesed'] == 10) | (df_drugs['Amount Range of Drug Possesed'] == 11), 'Severity of Crime'] = 3
df_drugs.loc[(df_drugs['Amount Range of Drug Possesed'] == 12) | (df_drugs['Amount Range of Drug Possesed'] == 13) | (df_drugs['Amount Range of Drug Possesed'] == 14), 'Severity of Crime'] = 2
df_drugs.loc[(df_drugs['Amount Range of Drug Possesed'] == 15) | (df_drugs['Amount Range of Drug Possesed'] == 16) | (df_drugs['Amount Range of Drug Possesed'] == 17), 'Severity of Crime'] = 1

# Criminal History Category:
# Current equal distribution
df_drugs.loc[:, 'Criminal History Category'] = np.random.choice(criminal_history_category, size=len(df_drugs))

# Guideline Sentence:
def est_sentencing(severity_n, history_n, months):
    e_size = df_drugs.loc[(df_drugs['Severity of Crime'] == severity_n) & (df_drugs['Criminal History Category'] == history_n), 'Guideline Sentence']
    df_drugs.loc[(df_drugs['Severity of Crime'] == severity_n) & (df_drugs['Criminal History Category'] == history_n), 'Guideline Sentence'] = np.random.choice(months, size=len(e_size))

severity_category = [(1,1,[i for i in range(0,7)]), 
                     (1,2,[i for i in range(0,7)]),
                     (1,3,[i for i in range(1,8)]),
                     (1,4,[i for i in range(3,10)]),
                     (1,5,[i for i in range(5,12)]),
                     (1,6,[i for i in range(7,14)]),
                    
                     (2,1,[i for i in range(7,14)]), 
                     (2,2,[i for i in range(10,17)]),
                     (2,3,[i for i in range(13,20)]),
                     (2,4,[i for i in range(18,25)]),
                     (2,5,[i for i in range(24,31)]),
                     (2,6,[i for i in range(27,35)]),
                    
                     (3,1,[i for i in range(27,35)]), 
                     (3,2,[i for i in range(30,39)]),
                     (3,3,[i for i in range(34,43)]),
                     (3,4,[i for i in range(42,53)]),
                     (3,5,[i for i in range(52,66)]),
                     (3,6,[i for i in range(58,73)]),
                    
                     (4,1,[i for i in range(58,73)]), 
                     (4,2,[i for i in range(65,81)]),
                     (4,3,[i for i in range(72,90)]),
                     (4,4,[i for i in range(85,107)]),
                     (4,5,[i for i in range(102,128)]),
                     (4,6,[i for i in range(111,138)]),
                     
                     (5,1,[i for i in range(123,155)]), 
                     (5,2,[i for i in range(138,174)]),
                     (5,3,[i for i in range(154,194)]),
                     (5,4,[i for i in range(172,236)]),
                     (5,5,[i for i in range(193,242)]),
                     (5,6,[i for i in range(215,270)])]

for sev in severity_category:
    est_sentencing(*sev)


df_drugs.loc[df_drugs['Severity of Crime'] == 6, 'Guideline Sentence'] = 'life'


# Actual Sentence:
def racial_bias(r, severness, prob):
    s_race = df_drugs.loc[df_drugs['Race'] == r]
    df_drugs.loc[df_drugs['Race'] == r, 'Bias'] = np.random.choice(severness, size=len(s_race), p=prob)

    df_drugs.loc[df_drugs['Race'] == r, 'Bias Severity'] = df_drugs.loc[df_drugs['Race'] == r,['Severity of Crime','Bias']].sum(axis=1)



racial_bias('Black', [i for i in range(-2,4)], [0.05,0.05,0.10,0.30,0.30,0.20])
racial_bias('Hispanic', [i for i in range(-2,3)], [0.05,0.05,0.20,0.35,0.35])
racial_bias('White', [i for i in range(-2,2)], [0.10,0.10,0.40,0.40])
racial_bias('Other', [i for i in range(-2,3)], [0.05,0.05,0.20,0.35,0.35])

df_drugs.loc[df_drugs['Bias Severity'] > 6, 'Bias Severity'] = 6
df_drugs.loc[df_drugs['Bias Severity'] < 0, 'Bias Severity'] = 0

# Populating Actual sentence
def given_sentencing(severity_n, history_n, months):
    e_size = df_drugs.loc[(df_drugs['Bias Severity'] == severity_n) & (df_drugs['Criminal History Category'] == history_n), 'Actual Sentence']
    df_drugs.loc[(df_drugs['Bias Severity'] == severity_n) & (df_drugs['Criminal History Category'] == history_n), 'Actual Sentence'] = np.random.choice(months, size=len(e_size))

for sev in severity_category:
    given_sentencing(*sev)


df_drugs.loc[df_drugs['Bias Severity'] == 6, 'Actual Sentence'] = 'life'

```
